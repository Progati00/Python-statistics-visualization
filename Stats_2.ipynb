# Q1. 1. Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`. After generating the list, find the following:      
# i) Write a Python function to calculate the mean of a given list of numbers. Create a function to find the median of a list of numbers.      
# ii) Develop a program to compute the mode of a list of integers.      
# iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights. 
# iv) Write a Python function to find the geometric mean of a list of positive numbers.      
# v) Create a program to calculate the harmonic mean of a list of values.      
# vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).      
# vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of outliers.

````python
import numpy as np
from scipy import stats

# Generate a list of 100 integers between 90 and 130
int_list = np.random.randint(90, 131, 100)

# (i) Function to calculate mean and median
def calculate_mean(numbers):
    return np.mean(numbers)

def calculate_median(numbers):
    return np.median(numbers)

mean_value = calculate_mean(int_list)
median_value = calculate_median(int_list)

# (ii) Function to calculate mode
def calculate_mode(numbers):
    mode_result = stats.mode(numbers)
    return mode_result.mode[0]  # mode() returns an object, extracting the first mode value

mode_value = calculate_mode(int_list)

# (iii) Function to calculate weighted mean
def calculate_weighted_mean(values, weights):
    return np.average(values, weights=weights)

# Example weights: create random weights between 1 and 5 for each value in the list
weights = np.random.randint(1, 6, len(int_list))
weighted_mean_value = calculate_weighted_mean(int_list, weights)

# (iv) Function to calculate geometric mean
def calculate_geometric_mean(numbers):
    return stats.gmean(numbers)

geometric_mean_value = calculate_geometric_mean(int_list)

# (v) Function to calculate harmonic mean
def calculate_harmonic_mean(numbers):
    return stats.hmean(numbers)

harmonic_mean_value = calculate_harmonic_mean(int_list)

# (vi) Function to calculate midrange
def calculate_midrange(numbers):
    return (np.min(numbers) + np.max(numbers)) / 2

midrange_value = calculate_midrange(int_list)

# (vii) Function to calculate trimmed mean (excluding outliers)
def calculate_trimmed_mean(numbers, proportion_to_cut):
    return stats.trim_mean(numbers, proportion_to_cut)

trimmed_mean_value = calculate_trimmed_mean(int_list, 0.1)  # Exclude 10% of outliers

# Displaying results
results = {
    "Mean": mean_value,
    "Median": median_value,
    "Mode": mode_value,
    "Weighted Mean": weighted_mean_value,
    "Geometric Mean": geometric_mean_value,
    "Harmonic Mean": harmonic_mean_value,
    "Midrange": midrange_value,
    "Trimmed Mean (10%)": trimmed_mean_value
}

for key, value in results.items():
    print(f"{key}: {value}")
````

int_list is generated using np. random.randint() to get 100 integers between 90 and 130.
Mode is calculated using stats.mode() from scipy library.
Trimmed mean excludes a specified percentage of outliers using stats.trim_mean().

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Generate a list of 500 integers between 200 and 300
int_list2 = np.random.randint(200, 301, 500)

# (i) Visualization of the given data
# 1. Frequency & Gaussian distribution
plt.figure(figsize=(12, 6))

# Frequency histogram
plt.subplot(1, 2, 1)
sns.histplot(int_list2, kde=False, color='blue', bins=20)
plt.title("Frequency Distribution")

# Gaussian distribution (Normal Distribution fit)
plt.subplot(1, 2, 2)
sns.histplot(int_list2, kde=True, color='red', bins=20, stat='density')
plt.title("Histogram with Gaussian Distribution (KDE)")

plt.show()

# (ii) Function to calculate the range of a given list of numbers
def calculate_range(numbers):
    return np.max(numbers) - np.min(numbers)

range_value = calculate_range(int_list2)

# (iii) Program to find the variance and standard deviation
def calculate_variance_and_std(numbers):
    variance = np.var(numbers, ddof=1)  # sample variance
    std_dev = np.std(numbers, ddof=1)  # sample standard deviation
    return variance, std_dev

variance_value, std_dev_value = calculate_variance_and_std(int_list2)

# (iv) Function to compute the interquartile range (IQR)
def calculate_IQR(numbers):
    Q1 = np.percentile(numbers, 25)
    Q3 = np.percentile(numbers, 75)
    return Q3 - Q1

IQR_value = calculate_IQR(int_list2)

# (v) Function to calculate the coefficient of variation
def calculate_coefficient_of_variation(numbers):
    mean = np.mean(numbers)
    std_dev = np.std(numbers, ddof=1)
    return std_dev / mean

coefficient_of_variation_value = calculate_coefficient_of_variation(int_list2)

# (vi) Function to find the mean absolute deviation (MAD)
def calculate_MAD(numbers):
    mean = np.mean(numbers)
    return np.mean(np.abs(numbers - mean))

MAD_value = calculate_MAD(int_list2)

# (vii) Program to calculate the quartile deviation
def calculate_quartile_deviation(numbers):
    Q1 = np.percentile(numbers, 25)
    Q3 = np.percentile(numbers, 75)
    return (Q3 - Q1) / 2

quartile_deviation_value = calculate_quartile_deviation(int_list2)

# (viii) Function to find the range-based coefficient of dispersion
def calculate_range_based_dispersion(numbers):
    range_value = calculate_range(numbers)
    mean = np.mean(numbers)
    return range_value / mean

range_based_dispersion_value = calculate_range_based_dispersion(int_list2)

# Results
results = {
    "Range": range_value,
    "Variance": variance_value,
    "Standard Deviation": std_dev_value,
    "Interquartile Range (IQR)": IQR_value,
    "Coefficient of Variation": coefficient_of_variation_value,
    "Mean Absolute Deviation (MAD)": MAD_value,
    "Quartile Deviation": quartile_deviation_value,
    "Range-based Coefficient of Dispersion": range_based_dispersion_value,
}

# Displaying results
for key, value in results.items():
    print(f"{key}: {value}")
````

# 3. Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.

````python
import numpy as np
import random
from scipy.stats import binom, poisson, norm

class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        return sum(v * p for v, p in zip(self.values, self.probabilities))

    def variance(self):
        mean = self.expected_value()
        return sum(p * (v - mean) ** 2 for v, p in zip(self.values, self.probabilities))

    
values = [1, 2, 3, 4, 5, 6]  # Possible outcomes of a fair dice
probabilities = [1/6] * 6    # Equal probabilities for each outcome
fair_dice = DiscreteRandomVariable(values, probabilities)
expected_value = fair_dice.expected_value()
variance = fair_dice.variance()
print(expected_value)
print(variance)
````

# 5. Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance.

````python
import numpy as np

def generate_samples(distribution, size, n=None, p=None, lam=None):
    if distribution == 'binomial':
        # Generate binomial distribution samples
        if n is None or p is None:
            raise ValueError("For binomial, both 'n' and 'p' must be provided")
        samples = np.random.binomial(n, p, size)
    elif distribution == 'poisson':
        # Generate poisson distribution samples
        if lam is None:
            raise ValueError("For poisson, 'lam' must be provided")
        samples = np.random.poisson(lam, size)
    else:
        raise ValueError("Unsupported distribution")

    # Calculate mean and variance
    mean = np.mean(samples)
    variance = np.var(samples)
    return mean, variance


# Binomial distribution (n=10, p=0.5, size=1000)
mean, variance = generate_samples('binomial', size=1000, n=10, p=0.5)
print(f"Binomial Distribution - Mean: {mean}, Variance: {variance}")

# Poisson distribution (lam=3, size=1000)
mean, variance = generate_samples('poisson', size=1000, lam=3)
print(f"Poisson Distribution - Mean: {mean}, Variance: {variance}")
````

# 6.  Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute the mean, variance, and standard deviation of the samples.

````python
import numpy as np

# Function to generate random numbers from a Gaussian distribution and compute mean, variance, and standard deviation
def generate_gaussian_samples(mean, std_dev, size):
    # Generate samples from a Gaussian distribution
    samples = np.random.normal(mean, std_dev, size)

    # Calculate mean, variance, and standard deviation
    calculated_mean = np.mean(samples)
    calculated_variance = np.var(samples)
    calculated_std_dev = np.std(samples)

    return calculated_mean, calculated_variance, calculated_std_dev

# Example usage:
# Parameters for the Gaussian distribution (mean=0, std_dev=1, size=1000)
mean = 0
std_dev = 1
size = 1000

# Generate samples and calculate statistics
mean_value, variance_value, std_dev_value = generate_gaussian_samples(mean, std_dev, size)


print(f"Generated Gaussian Distribution - Mean: {mean_value}, Variance: {variance_value}, Standard Deviation: {std_dev_value}")
````

# (v) Write a script to visualize the correlation between two specific columns in a Pandas DataFrame using scatter plots.

````python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, pearsonr

# Load the tips dataset
tips = sns.load_dataset('tips')

# (i) Function to calculate skewness
def calculate_skewness(column):
    return skew(column)

# Calculate skewness for 'total_bill' and 'tip'
total_bill_skewness = calculate_skewness(tips['total_bill'])
tip_skewness = calculate_skewness(tips['tip'])
print(f"Skewness of total_bill: {total_bill_skewness}")
print(f"Skewness of tip: {tip_skewness}")

# (ii) Program to determine the type of skewness
def determine_skewness(column_name, skew_value):
    if skew_value > 0:
        return f"{column_name} exhibits positive skewness."
    elif skew_value < 0:
        return f"{column_name} exhibits negative skewness."
    else:
        return f"{column_name} is approximately symmetric."

print(determine_skewness('total_bill', total_bill_skewness))
print(determine_skewness('tip', tip_skewness))

# (iii) Function to calculate covariance
def calculate_covariance(column1, column2):
    covariance_matrix = np.cov(column1, column2)
    return covariance_matrix[0, 1]

# Calculate covariance between 'total_bill' and 'tip'
covariance = calculate_covariance(tips['total_bill'], tips['tip'])
print(f"Covariance between total_bill and tip: {covariance}")

# (iv) Program to calculate Pearson correlation coefficient
def calculate_pearson_correlation(column1, column2):
    correlation, _ = pearsonr(column1, column2)
    return correlation

# Calculate Pearson correlation coefficient between 'total_bill' and 'tip'
pearson_corr = calculate_pearson_correlation(tips['total_bill'], tips['tip'])
print(f"Pearson correlation coefficient between total_bill and tip: {pearson_corr}")

# (v) Script to visualize correlation using scatter plot
def scatter_plot(data, column1, column2):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=column1, y=column2, data=data)
    plt.title(f"Scatter Plot: {column1} vs {column2}")
    plt.xlabel(column1)
    plt.ylabel(column2)
    plt.show()

# Visualize correlation between 'total_bill' and 'tip'
scatter_plot(tips, 'total_bill', 'tip')
````

# 8. Write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution.

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_normal_pdf(x, mean, std_dev):
    return stats.norm.pdf(x, loc=mean, scale=std_dev)

# Calculate PDF for a normal distribution with mean=0, std_dev=1
x_values = np.linspace(-5, 5, 100)
pdf_values = calculate_normal_pdf(x_values, 0, 1)
plt.plot(x_values, pdf_values)
plt.title('Normal Distribution: PDF')
plt.xlabel('x')
plt.ylabel('PDF')
plt.show()
````

# 11.  A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors who make a purchase). They collect data from the old and new layouts to compare. To generate the data use the following command:
```python
import numpy as np
# 50 purchases out of 1000 visitors
old_layout = np.array([1] * 50 + [0] * 950)
# 70 purchases out of 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)
```
# Apply z-test to find which layout is successful.

````python
import numpy as np
from scipy.stats import norm

old_layout = np.array([1]*50 + [0]*950)
new_layout = np.array([1]*70 + [0]*930)

old_rate, new_rate = np.mean(old_layout), np.mean(new_layout)
old_se = np.sqrt((old_rate * (1 - old_rate)) / len(old_layout))
new_se = np.sqrt((new_rate * (1 - new_rate)) / len(new_layout))

z = (new_rate - old_rate) / np.sqrt(old_se**2 + new_se**2)
p_value = 2 * (1 - norm.cdf(abs(z)))

print("New layout is better" if p_value < 0.05 else "No significant difference")
````

# 18.  An HR department wants to investigate if there's a gender-based salary gap within the company. Use the below code to generate synthetic data:
```python

male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)
```

# 21. A political analyst wants to determine if there is a significant association between age groups and voter preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify them into different age groups and candidate preferences. Perform a Chi-Square test to determine if there is a significant association between age groups and voter preferences. Use the below code to generate data:
```python
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+', '51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)
```

````python
import numpy as np
import pandas as pd
from scipy import stats

np.random.seed(0)

# Generate synthetic data
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

# Create a contingency table
data = pd.crosstab(age_groups, voter_preferences)

# Perform Chi-Square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(data)

# Check significance
print("Chi-Square statistic:", chi2_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("There is a significant association between age groups and voter preferences.")
else:
    print("There is no significant association between age groups and voter preferences.")
````

# 24. A company produces three different versions of a product: Standard, Premium, and Deluxe. The company wants to determine if there is a significant difference in customer satisfaction scores among the three product versions. They conducted a survey and collected customer satisfaction scores for each version from a random sample of customers. Perform an ANOVA test to determine if there is a significant difference in customer satisfaction scores. Use the following data:
```python
# Sample data: Customer satisfaction scores for each product version
standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]
```

````python
import numpy as np
import scipy.stats as stats

standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

# Perform ANOVA test
f_stat, p_value = stats.f_oneway(standard_scores, premium_scores, deluxe_scores)

# Check significance
print("F-statistic:", f_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("There is a significant difference in customer satisfaction scores among the product versions.")
else:
    print("There is no significant difference in customer satisfaction scores among the product versions.")
````
